```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
```

```python

iris = load_iris()
X = iris.data
```

```python

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python

dbscan_hard = DBSCAN(eps=0.5, min_samples=5)
y_hard = dbscan_hard.fit_predict(X)
```

```python

dbscan_hard = DBSCAN(eps=0.5, min_samples=5)
y_hardpca = dbscan_hard.fit_predict(X_pca)
```

```python

dbscan_soft = DBSCAN(eps=0.5, min_samples=5)
dbscan_soft.fit(X)
soft_labels = dbscan_soft.labels_
soft_scores = dbscan_soft.core_sample_indices_
```

```python

dbscan_soft = DBSCAN(eps=0.5, min_samples=5)
dbscan_soft.fit(X_pca)
soft_labelspca = dbscan_soft.labels_
soft_scores = dbscan_soft.core_sample_indices_
```

```python

plt.scatter(X[:, 0], X[:, 1], c=y_hard)
plt.title("Hard Clustering using DBSCAN on IRIS Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()

plt.scatter(X[:, 0], X[:, 1], c=y_hardpca)
plt.title("Hard Clustering using DBSCAN and PCA on IRIS Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()

plt.scatter(X[:, 0], X[:, 1], c=soft_labels, cmap='viridis')
plt.title("Soft Clustering using DBSCAN on IRIS Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()

plt.scatter(X[:, 0], X[:, 1], c=soft_labelspca, cmap='viridis')
plt.title("Soft Clustering using DBSCAN and PCA on IRIS Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```
